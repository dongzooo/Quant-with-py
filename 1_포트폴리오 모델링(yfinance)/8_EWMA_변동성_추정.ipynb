```python
# 패키지 임포트: numpy, pandas, matplotlib, seaborn, yfinance
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
sns.set()
```

```python
# 티커 객체
etf = yf.Ticker('SPY')

# 가격 데이터 다운로드
data = etf.history(start='2000-01-01', actions=False)

# 불필요한 칼럼 제거
data.drop(['Open', 'High', 'Low', 'Volume'], axis=1, inplace=True)
```

```python
# 수익률 계산
rets = data.pct_change().dropna()
```

### 지수가중이동평균 변동성 추정

```python
# 패러미터 = 0.2
alpha = 0.2

# EWMA 변동성
vols = rets.ewm(alpha=alpha).std() * np.sqrt(252)
```

```python
# 그래프 출력
vols.plot()
```

```python
alpha = 0.9
vols = rets.ewm(alpha=alpha).std() * np.sqrt(252)
```

```python
vols.plot()
```
