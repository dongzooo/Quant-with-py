```python
# 패키지 임포트
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
sns.set()
```

```python
# SPY ETF 티커 객체
etf = yf.Ticker('SPY')

# 가격 데이터 다운로드
data = etf.history(start='2000-01-01', actions=False)

# 불필요한 칼럼 제거
data.drop(['Open', 'High', 'Low', 'Volume'], inplace=True, axis=1)
```

```python
# 수익률 계산
rets = data.pct_change().fillna(0)
```

### 20일 롤링 변동성 추정

```python
# 패러미터 = 20일
param = 20

# 20일 롤링 변동성
vols = rets.rolling(param).std() * np.sqrt(252)
```

```python
# 결과 시각화
vols.plot()
```

```python
param = 60

vols_60 = rets.rolling(param).std() * np.sqrt(252)
```

```python
vols_60.plot()
```
